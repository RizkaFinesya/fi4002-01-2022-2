# Root Existence Interval

```python
import math
import matplotlib.pyplot as plt

def f(x):
    y = math.cos(x + 0.26 * math.pi)-0.37
    return y

x = [i * 0.06  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [3, 5]
yroot = [f(i) for i in xroot]

plt.grid()
plt.plot(x, y, 'b-', xroot, yroot, 'ro', markerfacecolor='none', ms=10, lw=2)
plt.show()

print("xroot =", xroot)
print("yroot =", yroot)
print("f(x1) f(x2) =", yroot[0] * yroot[1], "< -1")
print("Root existence:", True)
```
