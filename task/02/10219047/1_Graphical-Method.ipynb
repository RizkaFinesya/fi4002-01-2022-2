# Graphical Method

$f(x)=\cos(x+0.12\pi)-0.20$

Root finding using graphical method

```python
import math
import matplotlib.pyplot as plt

def f(x):
    y = math.cos(x + 0.12 * math.pi) - 0.20
    return y

x = [i * math.pi / 100  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

ans = math.pi / 4  + math.pi / 6
xroot = [1.00]
yroot = [0]

plt.grid()
plt.plot(x, y, xroot, yroot, 'r+', ms=10, lw=2)
plt.show()
```

```python
xroot = [1.00]
yroot = [0]

plt.grid()
plt.xlim(0.9, 1.1)
plt.ylim(-0.05, 0.05)
plt.plot(x, y, xroot, yroot, 'r+', ms=10, lw=2)
plt.show()
```
