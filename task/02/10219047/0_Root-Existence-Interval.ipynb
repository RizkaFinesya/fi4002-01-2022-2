# Root Existence Interval

$f(x)=\cos(x-0.12\pi)-0.20$

Check existence of root in an interval $x \in [0, 2]$

```python
import math
import matplotlib.pyplot as plt

def f(x):
    y = math.cos(x + 0.12 * math.pi)-0.20
    return y

x = [i * 0.04  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [0, 2]
yroot = [f(i) for i in xroot]

plt.grid()
plt.plot(x, y, 'b-', xroot, yroot, 'ro', markerfacecolor='none', ms=10, lw=2)
plt.show()

print("xroot =", xroot)
print("yroot =", yroot)
print("f(x1) f(x2) =", yroot[0] * yroot[1], "< -1")
print("Root existence:", True)
```
